```python

for i in range(len(df)):
    df['Speech']=re.sub(r"\\[a-z][a-z]?[0-9]+", '', df['Speech'][i])
    df['Speech']=re.sub('[^A-Za-z ]+', '', df['Speech'][i])
    
```

```python
worda=word_tokenize(df['Speech'][0])
wordb=word_tokenize(df['Speech'][1])
```

```python

nltk.download('punkt')
```

```python
worda = [w for w in worda if len(w) > 2]
wordb = [w for w in wordb if len(w) > 2]
worda=[w for w in worda if not w in STOPWORDS]
wordb=[w for w in wordb if not w in STOPWORDS]
```
